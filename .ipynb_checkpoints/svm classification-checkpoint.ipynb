```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC
from sklearn import metrics
```

# importing the dataset

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
X = df.iloc[:,0:2]
Y = df.iloc[:,2]
X.head()
```

# Splitting the dataset into the training and test set 

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
print("training:",x_train.shape)
print("testing:",x_test.shape)
```

# Feature Scaling

```python
sc_x =StandardScaler()
X_Train = sc_x.fit_transform(x_train)
X_Test  = sc_x.fit_transform(x_test)
```

# Fitting the model

# classify using linear kernel

```python
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_Train,y_train)
```

# predicting the test set result

```python
y_pred = classifier.predict(X_Test)
```

```python
y_pred
```

```python
print("Accuracy Score with Linear kernel : :: ",metrics.accuracy_score(y_test,y_pred))
```

# classify using  rbf kernel

```python
classifier = SVC(kernel='rbf', random_state=0)
classifier.fit(X_Train,y_train)
y_pred = classifier.predict(X_Test)
print("Accuracy Score with Linear kernel : :: ",metrics.accuracy_score(y_test,y_pred))
```

# classify using poly. kernel

```python
classifier = SVC(kernel='poly', degree=4)
classifier.fit(X_Train,y_train)
y_pred = classifier.predict(X_Test)
print("Accuracy Score with Linear kernel : :: ",metrics.accuracy_score(y_test,y_pred))
```

# scatting the data points of Training

```python
plt.scatter(X_Train[:,0],X_Train[:,1],c=y_train)
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.title("Training Data")
plt.show()
```

# Scatting the data points of testing

```python
plt.scatter(X_Test[:,0],X_Test[:,1],c=y_test)
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.title("Training Data")
plt.show()
```

# plot the hyperplane the is classify the data

```python
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(X_Train,y_train)
y_pred = classifier.predict(X_Test)
# print("Accuracy Score with Linear kernel : :: ",metrics.accuracy_score(y_test,y_pred))
# plot data points
plt.scatter(X_Test[:,0],X_Test[:,1],c=y_test)

# create the hyperplane 
w = classifier.coef_[0]
a = -w[0]/ w[1]
xx = np.linspace(-2.5,2.5)
yy=a*xx - (classifier.intercept_[0])/w[1]

# plot the hyperplane 
plt.plot(xx,yy)
plt.axis("off")
plt.show()
```
